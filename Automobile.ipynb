```python
import numpy as np
import pandas as pd
import xgboost
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from math import sqrt
from sklearn.metrics import mean_squared_error
```

```python
data = pd.read_csv('insurance.csv')
data
```

```python
data.dtypes
```

```python
X_cat = data.copy()
# print(X_cat) 
X_cat = data.select_dtypes(include=['object'])
X_enc = X_cat.copy()
X_enc
```

```python
#encoding 
# X_enc = pd.get_dummies(X_enc)
```

```python
X_enc
```

```python
data = data.drop(['sex','smoker','region'],axis=1)
```

```python
data
```

```python
X_enc = X_enc.apply(LabelEncoder().fit_transform)
X_enc
```

```python
data = pd.concat([data,X_enc],axis=1)
data
```

```python
y = data['charges'].astype('int')
x = data.drop(['charges'],axis=1) 
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.3)
```

```python
clf = xgboost.XGBRegressor()
clf.fit(X_train,Y_train)
y_testpred = clf.predict(X_test)
```

```python
rms = sqrt(mean_squared_error(Y_test,y_testpred))
print('RMSE: ',rms)
```
